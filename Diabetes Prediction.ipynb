**c . What can you interpret from the log probabilities and the predicted classes?**



Log probabilities represents the logarithmic values of the probabilities of all classes, in this case 0 and 1. It has two columns for classes 0 and 1. It shows a the values of both classes 0 and 1 for each sample. 0 corresponds to 'No' or in this case 'The person is not diabetic' and 1 corresponds to 'Yes' which in this translates to 'The person is diabetic'. Predicted classes give actual predicted class and not a log value of probabilities of both classes. If the value of probability of one class is more than the other, that class is assigned to the corresponding sample. For example, in the first sample of the training data, if we take the antilog of class 0, the value is 0.21. The antilog value for class 1 is 0.78. Thus the class for that sample becomes 1.
